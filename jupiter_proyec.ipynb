```python
#Extraer datos esenciales de un conjunto de datos y mostrarlos es una parte necesaria de la ciencia de datos; 
#por lo tanto, los individuos pueden tomar decisiones correctas basadas en los datos. En esta tarea, extraerá algunos indicadores económicos esenciales de algunos datos, 
#luego mostrará estos indicadores económicos en un Tablero. Tiene la opción de alojar el Tablero en IBM Object Storage.


#Datos de  "Activos de Información - ANCP-CCE - 2020" url: "https://www.datos.gov.co/Organismos-de-Control/Activos-de-Informaci-n-ANCP-CCE-2020/nanp-9aur"
#Conjunto de datos y archivo correspondiente al listado de Activos de Información de la Agencia Nacional de Contratación Pública - 
#Colombia Compra Eficiente para el año 2020


import pandas as pd
import matplotlib.pyplot as plt
import json
import requests
import sys
import numpy as np 

url = "https://www.datos.gov.co/resource/nanp-9aur.json"

#funcion que realiza peticion a api rest para traer datos json
def traer_datos(url_temp):
    r = requests.get(url_temp)
    files = r.json()
    return files

#funcion que se encarga de convertir lista a diccionario
def one_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}
    for dict_ in list_dict:
        for key, value in dict_.items():
            out_dict[key].append(value)
    return out_dict    

datos_dicc = one_dict(traer_datos(url))
#tratamiento de la informacion relevante 
colum_relevat = ["proceso","impacto_social_0_5_de_poblaci","impacto_econ_mico_pib_de","impacto_ambiental"]
#dicc = {name_process: [[impacto_social],[impacto_econ],[impacto_ambiental]]}

def extrac_datos_relev(date_dicc, col_relev):
    extrac_relev = {}
    for key, value in date_dicc.items():
        if key in col_relev:
            extrac_relev[key] = value
    return extrac_relev

datos_relev = extrac_datos_relev(datos_dicc,colum_relevat)

analisis_datos = {}
array_process = np.array(datos_relev["proceso"])
array_impac_social = np.array(datos_relev["impacto_social_0_5_de_poblaci"]) 
array_impac_econom = np.array(datos_relev["impacto_econ_mico_pib_de"]) 
array_impac_ambien = np.array(datos_relev["impacto_ambiental"]) 

for i in range(len(array_process)):
    if array_process[i] in analisis_datos.keys():
        var_temp = analisis_datos[array_process[i]]
        if (array_impac_social[i] == 'NO'):
            var_temp[0] = analisis_datos[array_process[i]][0]+1
        if (array_impac_econom[i] == 'NO'):
            var_temp[1] = analisis_datos[array_process[i]][1]+1
        if (array_impac_ambien[i] == 'NO'):
            var_temp[2] = analisis_datos[array_process[i]][2]+1
        analisis_datos[array_process[i]] = var_temp
    else:
        var_temp = [0,0,0]
        analisis_datos[array_process[i]] = var_temp
        if (array_impac_social[i] == 'NO'):
            var_temp[0] = analisis_datos[array_process[i]][0]+1
        if (array_impac_econom[i] == 'NO'):
            var_temp[1] = analisis_datos[array_process[i]][1]+1
        if (array_impac_ambien[i] == 'NO'):
            var_temp[2] = analisis_datos[array_process[i]][2]+1
        analisis_datos[array_process[i]] = var_temp

print("los procesos de contratacion y su NO influencia en la sociedad colombiana en los aspectos [social economica ambiental]")
df=pd.DataFrame(analisis_datos)
df.head()
```
